```python
import numpy as np
import pandas as pd

import ast
import nltk
from nltk.stem.porter import PorterStemmer

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
# Load Dataset
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
# Merge Datasets
movies = movies.merge(credits, on='title')
movies.head()
```

```python
# Data Cleaning
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]

movies.dropna(inplace=True)
```

```python
# Convert JSON-like columns

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# Extract Top Cast (first 3)
def convert_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L

movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
# Extract Director
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L

movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# Convert overview into list
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
# Remove Spaces (important for NLP)
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# Create Tags Column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# Final Dataset
new_df = movies[['movie_id','title','tags']]
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
# Stemming (important optimization)
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# Vectorization
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# Cosine Similarity
similarity = cosine_similarity(vectors)
```

```python
# Recommendation Function
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
# Test It
recommend("Avatar")
```

```python
# EDA
genre_counts = movies['genres'].explode().value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=genre_counts.values, y=genre_counts.index)
plt.title("Top Genres")
plt.show()
```

```python
# Movie Length Distribution
movies['overview_len'] = movies['overview'].apply(len)

plt.figure(figsize=(8,5))
sns.histplot(movies['overview_len'], bins=30)
plt.title("Overview Length Distribution")
plt.show()
```

```python
# Top Actors
cast_counts = movies['cast'].explode().value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=cast_counts.values, y=cast_counts.index)
plt.title("Top Actors")
plt.show()
```
